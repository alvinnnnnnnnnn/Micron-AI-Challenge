```python
test_df = filtered_run_df
test_df['time elapsed'] = (test_df['Time Stamp'] - test_df['Run Start Time']).dt.total_seconds()
test_df
```

```python
# Plot the chart
sensor_list = test_df["Sensor Name"].unique().tolist()

for sensor in sensor_list:
    temp_df = test_df[test_df["Sensor Name"]==sensor]
    fig = px.line(
        temp_df,
        x='time elapsed',
        y='Sensor Value',
        title=f'{sensor}',
        labels={'time elapsed': 'Time Elapsed', 'Sensor Value': 'Sensor Value'}
    )

    fig.show()
```
