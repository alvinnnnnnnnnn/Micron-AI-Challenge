```python
import pandas as pd
import numpy as np

# 1. Aggregate incoming_run_df with Step ID information
incoming_sensor_cols = [f'Sensor_{i}' for i in range(1, 42)]
incoming_agg = incoming_df_filtered_pivot.groupby(['Run ID', 'Step ID'])[incoming_sensor_cols].agg(['mean', 'max', 'min', 'std'])

# Properly flatten the multi-index columns
incoming_agg.columns = [f'incoming_{col[0]}_{col[1]}_step{step}' 
                       for col, step in zip(incoming_agg.columns, 
                                          incoming_agg.index.get_level_values('Step ID').repeat(len(incoming_sensor_cols)*4))]

# Pivot to get one row per Run ID
incoming_agg = incoming_agg.unstack('Step ID')
incoming_agg.columns = [f'{col[0]}_{col[1]}' for col in incoming_agg.columns]
incoming_agg = incoming_agg.reset_index()

# 2. Aggregate run_df with Step ID information
current_sensor_cols = [f'Sensor_{chr(i)}' for i in range(ord('A'), ord('O')+1)]
current_agg = run_df_filtered_pivot.groupby(['Run ID', 'Step ID'])[current_sensor_cols + ['Consumable Life']].agg(['mean', 'max', 'min', 'std'])

# Properly flatten the multi-index columns
current_agg.columns = [f'current_{col[0]}_{col[1]}_step{step}' 
                      for col, step in zip(current_agg.columns, 
                                         current_agg.index.get_level_values('Step ID').repeat(len(current_sensor_cols)*4 + 4))]

# Pivot to get one row per Run ID
current_agg = current_agg.unstack('Step ID')
current_agg.columns = [f'{col[0]}_{col[1]}' for col in current_agg.columns]
current_agg = current_agg.reset_index()

# 3. Merge with metrology data
final_df = metrology_df_filtered.merge(incoming_agg, on='Run ID', how='left')
final_df = final_df.merge(current_agg, on='Run ID', how='left')

# 4. Add tool metadata
tool_metadata = run_df[['Run ID', 'Tool ID', 'Process Step']].drop_duplicates()
final_df = final_df.merge(tool_metadata, on='Run ID', how='left')

# 5. Clean up column names (remove any duplicate prefixes)
final_df.columns = [col.replace('_stepstep_', '_step') for col in final_df.columns]

final_df
```
