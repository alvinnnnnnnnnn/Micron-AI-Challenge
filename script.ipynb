# Importing Libraries

```python
import pandas as pd 
import numpy as np

import warnings
warnings.filterwarnings("ignore")

import plotly.express as px
```

# Loading Dataset

```python
incoming_df = pd.read_parquet("train/incoming_run_data_1.parquet")
incoming_df_filtered = incoming_df[incoming_df["Run ID"] == "6515456b-05d4-5e6d-b42a-7652a8d671a4"]

incoming_df_filtered_pivot = incoming_df_filtered.pivot(index=["Tool ID", "Run Start Time", "Run End Time", "Run ID", 
                                                               "Process Step", "Time Stamp", "Step ID"], 
                                                        columns="Sensor Name", 
                                                        values="Sensor Value").reset_index()
incoming_df_filtered_pivot
```

```python
tmp_list = []

temp_grps = [(x[0], y) for x, y in incoming_df_filtered_pivot.groupby(by=["Step ID"])]
for temp_grp in temp_grps:
    temp_step_id = temp_grp[0]
    temp_df = temp_grp[1]
    temp_df = temp_df.reset_index()

    temp_tool_id = temp_df.loc[0, "Tool ID"]
    temp_run_start_time = temp_df.loc[0, "Run Start Time"]
    temp_run_end_time = temp_df.loc[0, "Run End Time"]
    temp_run_id = temp_df.loc[0, "Run ID"]
    temp_process_step = temp_df.loc[0, "Process Step"]

    # Sensor columns for incoming data are Sensor_1 to Sensor_41
    sensor_cols = [f"Sensor_{i}" for i in range(1, 42)]
    sensor_avgs = [temp_df[col].mean() for col in sensor_cols]

    temp_time_taken = (temp_df["Time Stamp"].max() - temp_df["Time Stamp"].min()).total_seconds()
    temp_time_taken = int(round(temp_time_taken))
   
    tmp_list.append([
        temp_process_step,
        temp_tool_id,
        temp_run_id,
        temp_run_start_time,
        temp_run_end_time,
        temp_time_taken,
        temp_step_id,
        *sensor_avgs  # unpack all sensor averages into the row
    ]) 

columns = [
    "Process Step",
    "Tool ID",
    "Run ID",
    "Run Start Time",
    "Run End Time",
    "Time Taken (Seconds)",
    "Step ID"
] + sensor_cols  # adds Sensor A to O
   
dataframe_incoming_filtered = pd.DataFrame(tmp_list, columns=columns) 
del temp_grps

dataframe_incoming_filtered = dataframe_incoming_filtered.sort_values(
    by=["Step ID"], 
    key=lambda x: x.str.extract(r'Step_(\d+)')[0].astype(int)
)
dataframe_incoming_filtered
```

```python
run_df = pd.read_parquet("train/run_data_1.parquet")
run_df_filtered = run_df[run_df["Run ID"] == "6515456b-05d4-5e6d-b42a-7652a8d671a4"]

run_df_filtered_pivot = run_df_filtered.pivot(index=["Tool ID", "Run Start Time", "Run End Time", "Run ID", 
                                                     "Process Step", "Consumable Life", "Time Stamp", "Step ID"], 
                                              columns="Sensor Name", 
                                              values="Sensor Value").reset_index()

run_df_filtered_pivot.to_parquet("run_data_pivot.parquet")

run_df_filtered_pivot
```

```python
tmp_list = []

temp_grps = [(x[0], y) for x, y in run_df_filtered_pivot.groupby(by=["Step ID"])]
for temp_grp in temp_grps:
    temp_step_id = temp_grp[0]
    temp_df = temp_grp[1]
    temp_df = temp_df.reset_index()

    temp_tool_id = temp_df.loc[0, "Tool ID"]
    temp_run_start_time = temp_df.loc[0, "Run Start Time"]
    temp_run_end_time = temp_df.loc[0, "Run End Time"]
    temp_run_id = temp_df.loc[0, "Run ID"]
    temp_process_step = temp_df.loc[0, "Process Step"]
    temp_consumable_life = temp_df.loc[0, "Consumable Life"]

    sensor_cols = [f"Sensor_{chr(i)}" for i in range(ord("A"), ord("O")+1)]
    sensor_avgs = [temp_df[col].mean() for col in sensor_cols]

    temp_time_taken = (temp_df["Time Stamp"].max() - temp_df["Time Stamp"].min()).total_seconds()
    temp_time_taken = int(round(temp_time_taken))
   
    tmp_list.append([
        temp_process_step,
        temp_tool_id,
        temp_run_id,
        temp_run_start_time,
        temp_run_end_time,
        temp_consumable_life,
        temp_time_taken,
        temp_step_id,
        *sensor_avgs  # unpack all sensor averages into the row
    ]) 

columns = [
    "Process Step",
    "Tool ID",
    "Run ID",
    "Run Start Time",
    "Run End Time",
    "Consumable Life",
    "Time Taken (Seconds)",
    "Step ID",
] + sensor_cols  # adds Sensor A to O
   
dataframe_run_filtered = pd.DataFrame(tmp_list, columns=columns) 
del temp_grps

# Sort by Step ID using the numeric value after 'Step_'
dataframe_run_filtered = dataframe_run_filtered.sort_values(
    by=["Step ID"], 
    key=lambda x: x.str.extract(r'Step_(\d+)')[0].astype(int)
)
dataframe_run_filtered
```

```python
dataframe_run_filtered_test = dataframe_run_filtered.drop(columns=["Time Taken (Seconds)"])
# Identify id columns and sensor columns
id_cols = [
    "Process Step", "Tool ID", "Run ID", "Run Start Time",
    "Run End Time", "Consumable Life"
]
step_col = "Step ID"
sensor_cols = [col for col in dataframe_run_filtered_test.columns if col not in id_cols + [step_col]]

# Melt to long format
df_long = dataframe_run_filtered_test.melt(id_vars = id_cols + [step_col],
                                      var_name = "Sensor Name",
                                      value_name = "Sensor Value"
)

df_long
```

```python
# Create sensor-step column names
df_long["Sensor_Step"] = df_long["Sensor"] + ";" + df_long[step_col].str.replace("Step_", "step_")

# Pivot to wide format
final_df = df_long.pivot_table(
    index=id_cols,
    columns="Sensor_Step",
    values="Value",
    aggfunc="first"
).reset_index()

# Reorder columns: id_cols first, then sorted sensor-step columns
sensor_step_cols = sorted([col for col in final_df.columns if ";" in str(col)])
final_df = final_df[id_cols + sensor_step_cols]

final_df
```
