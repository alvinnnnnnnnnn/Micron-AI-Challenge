```python
import pandas as pd
import os

# Define the folder path
train_folder = 'train'
```

```python
incoming_run_files = [f for f in os.listdir(train_folder) if 'incoming_run_data' in f]

# Read and concatenate all parquet files
incoming_dataframes = [pd.read_parquet(os.path.join(train_folder, file)) for file in incoming_run_files]
incoming_df = pd.concat(incoming_dataframes, ignore_index=True)

incoming_df.to_parquet('incoming_run_data.parquet', index=False)
```

```python
incoming_df = pd.read_parquet('incoming_run_data.parquet')
```

```python
run_files = [f for f in os.listdir(train_folder) if 'run_data' in f and 'incoming' not in f]

# Read and concatenate all parquet files
run_dataframes = [pd.read_parquet(os.path.join(train_folder, file)) for file in run_files]
run_df = pd.concat(run_dataframes, ignore_index=True)

run_df.to_parquet('run_data.parquet', index=False)
```

```python
run_df = pd.read_parquet('run_data.parquet')
```

```python
metrology_files = [f for f in os.listdir(train_folder) if 'metrology_data' in f]

# Read and concatenate all parquet files
metrology_dataframes = [pd.read_parquet(os.path.join(train_folder, file)) for file in metrology_files]
metrology_df = pd.concat(metrology_dataframes, ignore_index=True)

metrology_df.to_parquet('metrology_df.parquet', index=False)
```

```python
metrology_df = pd.read_parquet('metrology_df.parquet')
```
